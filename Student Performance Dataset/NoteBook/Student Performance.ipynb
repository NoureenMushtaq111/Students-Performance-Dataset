```python
import pandas as pd

# Load the student dataset
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
# Check shape and column types
print("Dataset Shape:", df.shape)
df.info()

# Check for any missing values
print("\nMissing values per column:\n", df.isnull().sum())
```

```python
# 1. Clean column names (replace spaces and slashes with underscores, convert to lowercase)
df.columns = df.columns.str.replace(' ', '_').str.replace('/', '_').str.lower()
```

```python

# 2. Clean text data inside columns (strip extra spaces and make lowercase)
text_cols = ['gender', 'race_ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course']
for col in text_cols:
    df[col] = df[col].str.strip().str.lower()

print("Cleaned Columns:", df.columns.tolist())
df.head()
```

```python
# Target columns (what we want to predict)
target_cols = ['math_score', 'reading_score', 'writing_score']
```

```python
# X contains demographics and student background details
X = df.drop(columns=target_cols)
```

```python
# y contains the actual test scores
y = df[target_cols]
```

```python
print("Features (X) shape:", X.shape)
print("Targets (y) shape:", y.shape)
```

```python
df.to_csv('Cleaned_Students_Performance.csv', index=False)
print("Success! Cleaned data saved.")
```
